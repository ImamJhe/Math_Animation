```python
import matplotlib.pyplot as plt
from IPython.display import Image, display

def Equation_to_Image(equation, name_file):
    formula = r'$ %s $' %equation
    fig = plt.figure()
    text = fig.text(0, 0, formula)
    
    # Saving the figure will render the text.
    dpi = 800
    fig.savefig(f'{name_file}.', dpi=dpi)
    
    # Now wecan work with text's bounding box.
    bbox = text.get_window_extent()
    width, height = bbox.size / float(dpi) + 0.005
    
    # Adjust the figure size so it can hold the entire text.
    fig.set_size_inches((width + .05, height + .05))
    
    # Adjust text's vertical position.
    dy = (bbox.ymin/float(dpi))/height
    text.set_position((0.03, -dy))
    
    # Save the adjusted text.
    fig.savefig(f'{name_file}.png', dpi=dpi)
    display(Image(f'{name_file}.png'))
```

```python
Equation = input("Equation :")
Name_file = input("Name File ")
Equation_to_Image(Equation, Name_file)
```
